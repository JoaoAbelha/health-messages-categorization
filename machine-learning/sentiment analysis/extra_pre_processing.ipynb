# WorNet lemmatization and POS tagging

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

#example text text = 'What can I say about this place. The staff of these restaurants is nice and the eggplant is not bad'

class Splitter(object):
    """
    split the document into sentences and tokenize each sentence
    """
    def __init__(self):
        self.splitter = nltk.data.load('tokenizers/punkt/english.pickle')
        self.tokenizer = nltk.tokenize.TreebankWordTokenizer()

    def split(self,text):
        """
        out : ['What', 'can', 'I', 'say', 'about', 'this', 'place', '.']
        """
        # split into single sentence
        sentences = self.splitter.tokenize(text)
        # tokenization in each sentences
        tokens = [self.tokenizer.tokenize(sent) for sent in sentences]
        return tokens


class LemmatizationWithPOSTagger(object):
    def __init__(self):
        pass
    def get_wordnet_pos(self,treebank_tag):
        """
        return WORDNET POS compliance to WORDENT lemmatization (a,n,r,v) 
        """
        if treebank_tag.startswith('J'):
            return wordnet.ADJ
        elif treebank_tag.startswith('V'):
            return wordnet.VERB
        elif treebank_tag.startswith('N'):
            return wordnet.NOUN
        elif treebank_tag.startswith('R'):
            return wordnet.ADV
        else:
            # As default pos in lemmatization is Noun
            return wordnet.NOUN

    def pos_tag(self,tokens):
        # find the pos tagginf for each tokens [('What', 'WP'), ('can', 'MD'), ('I', 'PRP') ....
        pos_tokens = [nltk.pos_tag(token) for token in tokens]

        # lemmatization using pos tagg   
        # convert into feature set of [('What', 'What', ['WP']), ('can', 'can', ['MD']), ... ie [original WORD, Lemmatized word, POS tag]
        pos_tokens = [ [(word, lemmatizer.lemmatize(word,self.get_wordnet_pos(pos_tag)), [pos_tag]) for (word,pos_tag) in pos] for pos in pos_tokens]
        return pos_tokens

lemmatizer = WordNetLemmatizer()
splitter = Splitter()
lemmatization_using_pos_tagger = LemmatizationWithPOSTagger()

#step 1 split document into sentence followed by tokenization

def lemmatise(text):
  tokens = splitter.split(text)
  #step 2 lemmatization using pos tagger 
  lemma_pos_token = lemmatization_using_pos_tagger.pos_tag(tokens)
  ans = ""
  for sentence in lemma_pos_token:
    for word in sentence:
      ans += word[1] + " "

  return ans[:-1]
```

```python
import re

def convert_emojis_to_word(text):
    for emot in EMOTICONS:
        text = re.sub(r'('+emot+')', "_".join(EMOTICONS[emot].replace(",","").replace(":","").split()), text)
    return text

def remove_emojis(text):
    for emot in EMOTICONS:
        text = re.sub(r'('+emot+')', "", text)
    return text   
```

```python
###########  stemming ####################

df['porter_sw_nltk_e'] = df['porter_sw_nltk'].apply(convert_emojis_to_word)
df['porter_sw_nltk_ne'] = df['porter_sw_nltk'].apply(remove_emojis)


df['porter_sw_nltkp_e'] = df['porter_sw_nltkp'].apply(convert_emojis_to_word)
df['porter_sw_nltkp_ne'] = df['porter_sw_nltkp'].apply(remove_emojis)


df['porter_sw_spacy_e'] = df['porter_sw_spacy'].apply(convert_emojis_to_word)
df['porter_sw_spacy_ne'] = df['porter_sw_spacy'].apply(remove_emojis)



df['porter_sw_spacyp_e'] = df['porter_sw_spacyp'].apply(convert_emojis_to_word)
df['porter_sw_spacyp_ne'] = df['porter_sw_spacyp'].apply(remove_emojis)



df['porter_no_sw_e'] = df['porter_no_sw'].apply(convert_emojis_to_word) # removing punct and no stopwords
df['porter_no_sw_ne'] = df['porter_no_sw'].apply(remove_emojis) # removing punct and no stopwords



df['porter_vanilla_e'] = df['porter_vanilla'].apply(convert_emojis_to_word) # no removing punct and no stopwords
df['porter_vanilla_ne'] = df['porter_vanilla'].apply(remove_emojis) # no removing punct and no stopwords

########### lemmatisation #################

df['lemma_sw_nltk_e'] = df['lemma_sw_nltk'].apply(convert_emojis_to_word)
df['lemma_sw_nltk_ne'] = df['lemma_sw_nltk'].apply(remove_emojis)


df['lemma_sw_nltkp_e'] = df['lemma_sw_nltkp'].apply(convert_emojis_to_word)
df['lemma_sw_nltkp_e'] = df['lemma_sw_nltkp'].apply(remove_emojis)


df['lemma_sw_spacy_e'] = df['lemma_sw_spacy'].apply(convert_emojis_to_word)
df['lemma_sw_spacy_ne'] = df['lemma_sw_spacy'].apply(remove_emojis)


df['lemma_sw_spacyp_e'] = df['lemma_sw_spacyp'].apply(convert_emojis_to_word)
df['lemma_sw_spacyp_ne'] = df['lemma_sw_spacyp'].apply(remove_emojis)


df['lemma_no_sw_e'] = df['lemma_no_sw'].apply(convert_emojis_to_word) # removing punct and no stopwords
df['lemma_no_sw_ne'] = df['lemma_no_sw'].apply(remove_emojis) # removing punct and no stopwords


df['lemma_vanilla_e'] = df['lemma_vanilla'].apply(convert_emojis_to_word) # no removing punct and no stopwords
df['lemma_vanilla_ne'] = df['lemma_vanilla'].apply(remove_emojis) # no removing punct and no stopwords
```
